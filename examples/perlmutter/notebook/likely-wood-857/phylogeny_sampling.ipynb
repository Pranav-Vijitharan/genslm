```python
from Bio import SeqIO
```

```python
generated_proteins = list(SeqIO.parse("likely-wood-857-mdh_gpt2_1024_translated.fasta", "fasta"))
```

```python
test_seqs = list(SeqIO.parse("/Users/mzvyagin/Documents/mdh_data/generate_data/full_mdh_fasta/test.fasta", "fasta"))
```

```python
test_protein_records = []
for s in test_seqs:
    protein_record = s.translate(to_stop=True)
    protein_record.name = s.name + "_translated"
    protein_record.description = s.description + "_translated"
    protein_record.id = s.id + "_translated"
    test_protein_records.append(protein_record)
```

```python
import random
random.seed(0)
```

```python
num_test_to_sample = 4000 - len(generated_proteins)
```

```python
test_seq_samples = random.sample(test_protein_records, num_test_to_sample)
```
